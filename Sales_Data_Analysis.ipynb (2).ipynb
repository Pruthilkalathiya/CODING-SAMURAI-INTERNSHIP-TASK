```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sales_data.csv")
df.head()

df.info()
df.describe()

df['Total'] = df['Quantity'] * df['Price']
print("Total Sales:", df['Total'].sum())

top_product = df.groupby('Product')['Total'].sum().sort_values(ascending=False)
print(top_product)

sns.barplot(x=top_product.index, y=top_product.values)
plt.title("Total Sales by Product")
plt.show()

top_product.plot.pie(autopct='%1.1f%%')
plt.title("Sales Distribution")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('sales_data.csv')
df.head()

df['Total'] = df['Quantity'] * df['Price']
df.head()

print(" Total Sales:", df['Total'].sum())

product_sales = df.groupby('Product')['Total'].sum().sort_values(ascending=False)
print(product_sales)

sns.barplot(x=product_sales.index, y=product_sales.values)
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()

product_sales.plot.pie(autopct='%1.1f%%', figsize=(5,5))
plt.title("Sales Distribution by Product")
plt.ylabel("")  # removes text label on side
plt.show()

df_numeric = df.select_dtypes(include=['int64', 'float64'])
df_numeric.corr()
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
```
